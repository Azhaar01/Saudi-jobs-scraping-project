### WEB SCRAPING

Importing Libraries & Methods

```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import pandas as pd
```

```python
# Write the header in the file
file = open('C:/Users/Azhaa/Downloads/Wazzuf-Saudi-Arabia.csv', 'a')
header = "job_title,company_name,location,job_type,job_location_type,experence_level,experence_years,job_category,skills_required\n"
file.write(header)


for i in range(0, 6):
    # importing the URL
    URL = f'https://wuzzuf.net/saudi/search/jobs/?a=navbg&filters%5Bcountry%5D%5B0%5D=Saudi%20Arabia&q=saudi%20arabia&start={i}'
    
    # Create a Client-based Request to get the URL
    client = urlopen(URL)
    
    # Getting the HTML Code of The Full Page
    HTML = client.read()
    
    # Closing The Request
    client.close()
    
    # Creating an HTML Parser Using BeatuifulSoup
    soup = bs(HTML, 'html.parser')
    
    # Creating a Container for the Needed Data
    containers = soup.find_all('div', {'class':'css-1gatmva e1v1l3u10'})
    
    # Accessing Page Elements
    for container in containers:
        jtitle = container.findAll('h2', {'class': 'css-m604qf'}) # اعطيه كلاس اتش تو
        job_title = jtitle[0].text.strip()

        company = container.findAll('a' , {'class': 'css-17s97q8'})
        company_name = company[0].text.strip()

        location = container.findAll('span' , {'class': 'css-5wys0k'})
        job_location = location[0].text.strip()

        jtype = container.findAll('a' , {'class': 'css-n2jc4m'})
        job_type = jtype[0].text.strip()

        job_type2 = container.findAll('span' , {'class': 'css-o1vzmt eoyjyou0'})
        job_location_type = job_type2[0].text.strip()

        Explevel = container.findAll('a', {'class' : 'css-o171kl'})
        Experence_level = Explevel[1].text.strip()

        Expyears = container.findAll('span')[3]
        Experence_years = Expyears.text[2:].strip()

        jcategory = container.findAll('a', {'class' : 'css-o171kl'})
        job_category = jcategory[2].text.strip()

        Skills = container.findAll('a', {'class' : 'css-5x9pm1'})
        skills_required = ''
        for skill in range(len(Skills)):
            skills_required = skills_required + Skills[skill].text.strip()
        
        file.write(job_title + ',' + company_name + ',' + f'"{job_location}"' + ',' + job_type + ',' + job_location_type + ',' + Experence_level + ',' + Experence_years + ',' +job_category + ',' + f'"{skills_required}"' + '\n')

file.close()
```
